```python
from gen.gen_utilities import rotVecToQuat, quatToRotMat, rotMatToQuat, rotTensor, quatToRotVec
import numpy as np
```

```python
# Test 0. - Indentity rotation
rotVec = np.array([0,0,0])
resultQat = np.array([1,0,0,0])
resultRotMat = np.eye(3)

quat = rotVecToQuat(rotVec)
vec = quatToRotVec(quat)
rotMat = quatToRotMat(quat)

print("Test 0 - Identity rotation")
print('Quaternion:', quat)
print('Rotation Matrix:', rotMat)
print('Quaternion from rotation matrix:', rotMatToQuat(rotMat))
print('Vector from quaternion:', vec)

error = resultRotMat.T @ rotMat
print('Error in rotation matrix:', error)
```

```python
a = rotTensor(rotVec=rotVec)
print('Rotation tensor from Scipy rotation function:', a)
if np.isclose(a, rotMat).all():
    print('SciPy rotation tensor matches rotation matrix.')
```
