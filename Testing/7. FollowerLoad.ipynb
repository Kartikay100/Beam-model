This file tests if the modified natural BC is correct to model follower force condition. It checks if the force is orthogonal to the elment. A beam of length 100cm, 10 elements/2 nodes is assumed. Analysis is carreid out on the last element of beam, after 1st iteration. Check output JSON files 7a and 7b in the folder.

```python
# Inputs

import numpy as np
a1 = np.array([ # first node of element
      0.807857698013967,
      0.0,
      89.99999999999997
    ])
a2 = np.array([ # second node of element
      0.9500006167529831,
      0.0,
      99.99999999999997
    ])
rotation = np.array([
      [
        0.9998979609194895,
        0.0,
        0.014285228351247
      ],
      [
        0.0,
        1.0,
        0.0
      ],
      [
        -0.014285228351247,
        0.0,
        0.9998979609194895
      ]])
force = np.array([10000,0,0])
```

```python
# element vector
a = a2-a1
# rotate the NBC by the same amount as the element.
rotatedForce = rotation @ force
```

```python
# normal vectors in 2 directions, force and element
dirA = a/np.linalg.norm(a)
dirRotatedForce = rotatedForce/np.linalg.norm(rotatedForce) # norm=10_000 N, same as norm of force
```

```python
# dot product between element vector and rotated force vector should be zero.
dirA @ dirRotatedForce
```

close to zero.
