```python
# check for ill-conditioning by liinearly interpolating between two rotation vectors
from scipy.spatial.transform import Rotation as R
import numpy as np
ep = 1E-20
a = np.pi-ep
r1 = np.array([0,a,0])
r2 = np.array([0,-a,0])
interp = np.array([0.5,0.5])
RotN1 = R.from_rotvec(r1).as_matrix()
RotN2 = R.from_rotvec(r2).as_matrix()
Rotref = RotN1
RotPhy2 = R.from_matrix(Rotref.T @ RotN2).as_rotvec()
RotPhy1 = np.array([0,0,0])
rotation = np.array([RotPhy1,RotPhy2])
interpVec = np.dot(interp, rotation)
R.from_matrix(Rotref @ R.from_rotvec(interpVec).as_matrix()).as_rotvec()
```
